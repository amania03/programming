```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Q2: Plotting with Matplotlib and NumPy

# Part 1 to 4: Plotting Natural Log and Exponential Functions
x = np.linspace(0.01, 1, 50)  # Create an array of 50 points from 0.01 to 1
y = np.log(x)  # Compute natural logarithm of x
z = np.exp(x)  # Compute exponential of x

# Scatter plot of Natural Log and Exponential Functions
plt.scatter(x, y, label='Natural Log')  # Scatter plot for natural log
plt.scatter(x, z, label='Exponential')  # Scatter plot for exponential
plt.xlabel("X")  # Label for x-axis
plt.ylabel("Y")  # Label for y-axis
plt.legend(loc='lower right')  # Add a legend in the lower right
plt.title('Natural Log and Exponential Functions')  # Title of the plot
plt.show()  # Display the plot

# Part 8 to 12: Sine and Cosine Functions
x = np.linspace(-6.5, 6.5, 100)  # Create an array of 100 points from -6.5 to 6.5
y = np.sin(x)  # Compute sine of x
z = np.cos(x)  # Compute cosine of x

# Line plot of Sine and Cosine Functions
plt.plot(x, y, label='Sine')  # Line plot for sine
plt.plot(x, z, label='Cosine')  # Line plot for cosine
plt.xlabel("X")  # Label for x-axis
plt.ylabel("Y")  # Label for y-axis
plt.legend(loc='lower left')  # Add a legend in the lower left
plt.title('Sine and Cosine')  # Title of the plot
plt.show()  # Display the plot

# Q3: Writing Statistical Functions

# Generating random values for testing
np.random.seed(100)  # Set the seed for reproducibility
rho, sigma_x, sigma_y = -.4, 3, 2  # Variance-Covariance Parameters
vcv = np.array([[sigma_x**2, rho*sigma_x*sigma_y],  # Variance-Covariance Matrix
                [rho*sigma_x*sigma_y, sigma_y**2]])
mu = np.array([-1, 2])  # Population averages
sample = np.random.multivariate_normal(mu, vcv, 200)  # Generate multivariate normal draws
x = sample[:, 0]  # Extract first variable
y = sample[:, 1]  # Extract second variable

# Part 1: Function to Compute Mean
def mean(x):
    return np.sum(x) / len(x)

# Part 2: Function to Compute Standard Deviation
def sd(x):
    m = mean(x)
    return np.sqrt(np.sum((x - m) ** 2) / (len(x) - 1))

# Part 3: Function to Compute Z-Scores
def zscore(x):
    return (x - mean(x)) / sd(x)

# Part 4: Function to Compute Covariance
def cov(x, y):
    return np.sum((x - mean(x)) * (y - mean(y))) / (len(x) - 1)

# Part 5: Function to Compute Correlation Coefficient
def corr(x, y):
    return cov(x, y) / (sd(x) * sd(y))

# Print results for verification
print('Sample Mean:', mean(x))
print('Sample Standard Deviation:', sd(x))
print('Z-scores:', zscore(x)[:10])  # Print first 10 z-scores
print('Covariance:', cov(x, y))
print('Correlation:', corr(x, y))

# Q4: Optimization Function

# Function to maximize a given function on a grid
def maximize_function(f, x_min, x_max, steps):
    x = np.linspace(x_min, x_max, steps)  # Create a grid of points
    f_values = f(x)  # Compute function values
    max_value = np.max(f_values)  # Find the maximum value
    max_x = x[np.argmax(f_values)]  # Find the x value corresponding to the maximum
    return {'max_value': max_value, 'max_x': max_x}

# Example usage of the maximize_function with a quadratic function
result = maximize_function(lambda x: -(x - 2)**2 + 1, -10, 10, 1000)
print('Maximizer:', result['max_x'])
print('Maximum Value:', result['max_value'])
#Assistance used from Stackoverflow and AI
```
